```python
total_cmnt_by_hour = {} # "9":520
no_post_by_hour = {}

for post in ask_posts:
    hr = post[-1].strftime("%H")
    cmnt = int(post[4])
    if hr not in total_cmnt_by_hour:
        total_cmnt_by_hour[hr] = cmnt
        no_post_by_hour[hr]= 1        
    else:
        total_cmnt_by_hour[hr] += cmnt
        no_post_by_hour[hr]+= 1  
```

```python
avg_cmnt_hr = {}

for hr in total_cmnt_by_hour:
    cmnt = total_cmnt_by_hour[hr]
    n_post = no_post_by_hour[hr]
    avg_cmnts = cmnt/n_post
    avg_cmnt_hr[hr] = round(avg_cmnts, 2)
```

```python
print(avg_cmnt_hr)
```

```python
avg_cmnt_hrlst = []

for h,c in avg_cmnt_hr.items():
    data = [h,c]
    avg_cmnt_hrlst.append(data)
```
